Function for convertion time into minutes

```python
def convertion_time(t):
    h, m, s = t.split(':')
    return float(h) * 60 + float(m)  + float(s) / 60
```

Filtarting the table only for retail purchases

```python
import csv

all_m, passed_m, bad_m = set(), set(), set()
results_by_date = {}
file = "file.csv"
with open(file, "r") as f:
    reader = csv.DictReader(f)
    with open("time_on_task.csv", "w", newline = "") as outfile:
        fieldnames = ["date", "moderator", "operation_time_minutes"]
        writer = csv.DictWriter(outfile, fieldnames = fieldnames)
        writer.writeheader()

        for row in reader:
            if row.get("team").strip() == "retail":
                date = row.get("start_time").split()[0]
                start_minute = convertion_time(row.get("start_time").split()[1])
                finish_minute = convertion_time(row.get("finish_time").split()[1])
                operation_time = round(abs(finish_minute - start_minute), 2)

                m = row.get("moderator")
                all_m.add(m)
                if operation_time <= 5:
                    print(f"{m} passed with time: {operation_time}")
                    passed_m.add(m)
                else:
                    bad_m.add(m)

                if date not in results_by_date:
                    results_by_date[date] = []
                results_by_date[date].append({"moderator": m, "operation_time_minutes": operation_time})

        for date, actions in sorted(results_by_date.items()):
            for action in actions:
                writer.writerow({"date": date, "moderator": action["moderator"], "operation_time_minutes": action["operation_time_minutes"]})
```

Printing those results

```python
print("Moderators, whose tasks time > 5 min: ")
for i in sorted(bad_m):
    print(i, end=" ")

print("\nSuccesfully passed moderators:")
for j in sorted(passed_m):
    print(j, end=" ")

print("\nAll moderators existed:")
for k in sorted(all_m):
    print(k, end=" ")
```
